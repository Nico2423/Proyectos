Vamos a averiguar cuál es el producto más vendido

```python
df.groupby('Producto')['Cantidad'].sum().sort_values(ascending=False).head(1)
```

1. El producto más vendido son los alimentos

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
```

```python
type(df['Fecha'][0])
```

```python
meses = []
for f in df['Fecha']:
    meses.append(f.month)
df['Meses'] = meses
```

```python
df
```

```python
df.groupby('Meses')['Total Venta'].sum().sort_values(ascending=False).head(1)
```

El mes en que más ingresos se registraron fue en agosto


```python
df.groupby('Producto')['Total Venta'].sum()
```

Aunque el producto que más se vendió en cuanto a cantidad fueron los alimentos, claramente el que dejó un mayor ingreso fueron los Electrónicos
